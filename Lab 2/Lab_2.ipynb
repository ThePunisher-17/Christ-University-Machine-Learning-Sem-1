# Feature Engineering

    Creating Rating column

    Inferance: The following function will return value between 0-2.5 if its not-satisfied and 2.5-5 if satisfied 

```python
from random import uniform
def random_rating(satisfaction):
    rating_stack = []
    for rating in satisfaction:
        if rating == 'satisfied':
            rating_stack.append(round(uniform(2.5, 5), 1))
        else:
            rating_stack.append(round(uniform(0, 2.5), 1))

    return rating_stack
```

    Inferance: Creating new column called Rating which uses above function creating new column

```python
rating = random_rating(df['satisfaction'])
df["Rating"] = rating
df.head()
```

    Creating Repeat Purchase

    Inferance: The following function will return 1 if customer rating is above 3 else it'll return 0 

```python
def random_repeat_purchase(ratings):
    repeat_purchase_stack = []
    for rating in ratings:
        if rating >= 3:
            repeat_purchase_stack.append(1)
        else:
            repeat_purchase_stack.append(0)
    

    return repeat_purchase_stack
```

    Inferance: The following lines will create new column called 'Repeat Purchaser' by using above function

```python
repeat_purchaser = random_repeat_purchase(df['Rating'])
df["Repeat Purchaser"] = repeat_purchaser
df.head()
```
